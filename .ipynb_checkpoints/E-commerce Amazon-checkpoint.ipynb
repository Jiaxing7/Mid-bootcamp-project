## change to numerical
### 100 Indische Rupie = 1.13 Euro  all Prices change to Euro

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹', '')
df['discounted_price'] = df['discounted_price'].str.replace(',', '')
df['discounted_price'] = df['discounted_price'].astype(float)
df['discounted_price_euro'] = round(df['discounted_price']/88.81,2)
```

```python
df['actual_price'] = df['actual_price'].str.replace('₹', '')
df['actual_price'] = df['actual_price'].str.replace(',', '')
df['actual_price'] = df['actual_price'].astype(float)
df['actual_price_euro'] = round(df['actual_price']/88.81,2)
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')
df['discount_percentage'] = df['discount_percentage'].astype(float)
df['discount_percentage'] = df['discount_percentage']/100
df.head()
```

```python
df['rating'].value_counts()
```

```python
df = df.loc[df['rating'] != '|']
df['rating'] = df['rating'].astype(float)
```

```python
df['rating_count'] = df['rating_count'].str.replace(',', '')
df['rating_count'] = df['rating_count'].astype(float)
df.head()
```

```python
df.dtypes
```

## deal with NaN values

# regression model

```python
df_cat = df[['category1']]
df_num = df[['discounted_price_euro','actual_price_euro','discount_percentage', 'rating', 'rating_count']]
print(df_cat.shape)
print(df_num.shape)
```

```python
df_full = pd.concat([df_cat,df_num], axis = 1)
df_full.shape
```

```python
df_full.head()
```

### X-y Split  (set rating_count as target)

```python
y = df_full['rating_count']
X = df_full.drop(['rating_count'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
X_train_num = X_train.select_dtypes('number')
X_train_cat = X_train.select_dtypes(object)

X_test_num = X_test.select_dtypes('number')
X_test_cat = X_test.select_dtypes(object)
```

### MinMaxScaler

```python
transformer = MinMaxScaler().fit(X_train_num)
# Scaling
X_train_scaled = pd.DataFrame(transformer.transform(X_train_num), columns=X_train_num.columns)
X_test_scaled = pd.DataFrame(transformer.transform(X_test_num), columns=X_test_num.columns)
```

### OneHot Encoding

```python
# Fitting the encoder
encoder = OneHotEncoder(drop='first', handle_unknown = 'ignore').fit(X_train_cat)
column_name = encoder.get_feature_names_out(X_train_cat.columns)
# Encoding
X_train_hot = pd.DataFrame(encoder.transform(X_train_cat).toarray(), columns = column_name)
X_test_hot = pd.DataFrame(encoder.transform(X_test_cat).toarray(), columns = column_name)
```

```python
display(X_train_hot.head(1))
```

```python
X_train_processed = pd.concat([X_train_scaled,X_train_hot], axis = 1)
X_test_processed = pd.concat([X_test_scaled,X_test_hot], axis = 1)
```

###  linear regression to see whether we are getting good results.

```python
lm = linear_model.LinearRegression()
```

```python
# Fitting the model
lm.fit(X_train_processed,y_train)
# Making predictions
predictions = lm.predict(X_test_processed)
```

```python
# Training Data
# R2-Score
predictions = lm.predict(X_train_processed)
r2score = round(r2_score(y_train, predictions), 2)
# Mean Square Error
mse= round(mean_squared_error(y_train,predictions))
# Root Mean Square Error
rmse = round(np.sqrt(mse), 2)
# Mean Absolute Error
mae = round(mean_absolute_error(y_train, predictions), 2)
# Test Data
# R2-Score
predictions = lm.predict(X_test_processed)
r2scoret = round(r2_score(y_test, predictions), 2)
# Mean Square Error
mset= round(mean_squared_error(y_test,predictions), 2)
# Root Mean Square Error
rmset = round(np.sqrt(mset), 2)
```

```python
display(pd.DataFrame({'ValType': ['Train', 'Test'], 'R2-Score': [r2score, r2scoret], 'MSE': [mse, mset] , 'RMSE': [rmse, rmset]}))
```

#### it's not good to be a linear regression
